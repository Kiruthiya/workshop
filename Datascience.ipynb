```python
import numpy as np
import pandas as pd
np.random.seed(45)
data = {
    "x" : np.arange(1,10),
    "y" : np.random.randint(12,80,size=9)
}
df=pd.DataFrame(data)
print(df)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(6,6))
plt.plot(df["x"],df["y"])
plt.xlabel("kir")
plt.ylabel("rik")
plt.title("plot")
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(6,6))
plt.plot(df["x"],df["y"],marker='*')
plt.xlabel("kir")
plt.ylabel("rik")
plt.title("plot")
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x=np.arange(1,11)
y=np.random.randint(10,100,size=10)
data=np.random.randn(100)
print(x)
print(y)
```

```python
plt.figure(figsize=(6,6))
plt.plot(x,y,linestyle='--',color='green')
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("Line plot")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(6,4))
plt.scatter(x,y,color = 'purple')
plt.xlabel("x")
plt.ylabel("y")
plt.title("scatter plot")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(6,4))
plt.bar(x,y,color = 'purple')
plt.xlabel("categories")
plt.ylabel("values")
plt.title("bar plot")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(6,4))
plt.barh(x,y,color = 'purple')
plt.xlabel("categories")
plt.ylabel("values")
plt.title("Horizontal Bar plot")
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(0, 11)
y = x ** 2

plt.plot(x, y, marker='o')
plt.title("Line Graph of y = x")
plt.xlabel("x values")
plt.ylabel("y values")
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

products = ['A', 'B', 'C', 'D']
sales = [40, 70, 30, 90]

plt.bar(products, sales, color=['skyblue', 'orange', 'green', 'red'])
plt.title("Product Sales Comparison")
plt.xlabel("Products")
plt.ylabel("Sales")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)

plt.figure(figsize=(8, 8))

plt.subplot(2, 2, 1)
plt.plot(x, np.sin(x))
plt.title("Line Plot")

plt.subplot(2, 2, 2)
plt.bar(['A', 'B', 'C'], [5, 7, 3])
plt.title("Bar Chart")

plt.subplot(2, 2, 3)
plt.scatter(x, np.random.randn(100))
plt.title("Scatter Plot")

plt.subplot(2, 2, 4)
plt.hist(np.random.randn(100), bins=10)
plt.title("Histogram")

plt.tight_layout()
plt.show()
```
